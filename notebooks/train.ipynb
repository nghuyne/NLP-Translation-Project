```python
tokenizer_en, tokenizer_fr = None, None  # Vì không dùng spaCy nữa

src_stoi, src_itos = build_vocab([p[0] for p in train_pairs])
trg_stoi, trg_itos = build_vocab([p[1] for p in train_pairs])
```

```python
encoder = Encoder(
    vocab_size=len(src_stoi),
    embed_size=256,
    hidden_size=512,
    num_layers=2,
    dropout=0.3
)

decoder = Decoder(
    vocab_size=len(trg_stoi),
    embed_size=256,
    hidden_size=512,
    num_layers=2,
    dropout=0.3
)

model = Seq2Seq(encoder, decoder, device).to(device)
```
