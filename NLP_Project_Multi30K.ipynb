```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence, pack_padded_sequence, pad_packed_sequence
from torch.utils.data import Dataset, DataLoader

import spacy
import matplotlib.pyplot as plt
from sacrebleu import corpus_bleu

import random
import math
import time
```

```python
spacy_en = spacy.load("en_core_web_sm")
spacy_fr = spacy.load("fr_core_news_sm")

def tokenize_en(text):
    return [tok.text.lower() for tok in spacy_en.tokenizer(text)]

def tokenize_fr(text):
    return [tok.text.lower() for tok in spacy_fr.tokenizer(text)]
```

```python
def load_parallel(src_path, trg_path, max_samples=None):
    src_lines = open(src_path, encoding="utf-8").read().strip().split("\n")
    trg_lines = open(trg_path, encoding="utf-8").read().strip().split("\n")
    pairs = list(zip(src_lines, trg_lines))
    if max_samples:
        pairs = pairs[:max_samples]
    return pairs

train_pairs = load_parallel("data/train.en", "data/train.fr")
val_pairs   = load_parallel("data/val.en", "data/val.fr")
test_pairs  = load_parallel("data/test.en", "data/test.fr")

print("Train:", len(train_pairs))
print("Val:", len(val_pairs))
print("Test:", len(test_pairs))
```

```python
PAD = "<pad>"
UNK = "<unk>"
SOS = "<sos>"
EOS = "<eos>"

def build_vocab(sentences, tokenizer, min_freq=2, max_vocab=10000):
    freq = {}
    for sent in sentences:
        for tok in tokenizer(sent):
            freq[tok] = freq.get(tok, 0) + 1

    vocab = [PAD, UNK, SOS, EOS]
    vocab += sorted([w for w, c in freq.items() if c >= min_freq])[:max_vocab]

    stoi = {w: i for i, w in enumerate(vocab)}
    itos = {i: w for w, i in stoi.items()}

    return stoi, itos

src_stoi, src_itos = build_vocab([p[0] for p in train_pairs], tokenize_en)
trg_stoi, trg_itos = build_vocab([p[1] for p in train_pairs], tokenize_fr)

print("SRC vocab size:", len(src_stoi))
print("TRG vocab size:", len(trg_stoi))
```

```python
class TranslationDataset(Dataset):
    def __init__(self, pairs):
        self.pairs = pairs

    def __len__(self):
        return len(self.pairs)

    def numericalize(self, text, stoi, tokenizer):
        return [stoi.get(tok, stoi[UNK]) for tok in tokenizer(text)]

    def __getitem__(self, idx):
        src, trg = self.pairs[idx]

        src_ids = [src_stoi[SOS]] + self.numericalize(src, src_stoi, tokenize_en) + [src_stoi[EOS]]
        trg_ids = [trg_stoi[SOS]] + self.numericalize(trg, trg_stoi, tokenize_fr) + [trg_stoi[EOS]]

        return torch.tensor(src_ids), torch.tensor(trg_ids)


def collate_fn(batch):
    src_batch, trg_batch = zip(*batch)

    src_lens = torch.tensor([len(x) for x in src_batch])
    trg_lens = torch.tensor([len(x) for x in trg_batch])

    src_pad = pad_sequence(src_batch, padding_value=src_stoi[PAD])
    trg_pad = pad_sequence(trg_batch, padding_value=trg_stoi[PAD])

    return src_pad, src_lens, trg_pad, trg_lens


train_loader = DataLoader(TranslationDataset(train_pairs), batch_size=32,
                          shuffle=True, collate_fn=collate_fn)

val_loader = DataLoader(TranslationDataset(val_pairs), batch_size=32,
                        shuffle=False, collate_fn=collate_fn)
```

```python
class Encoder(nn.Module):
    def __init__(self, vocab, emb_dim, hid_dim, n_layers, dropout):
        super().__init__()
        self.embedding = nn.Embedding(vocab, emb_dim)
        self.lstm = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout)

    def forward(self, src, src_len):
        embedded = self.embedding(src)
        packed = pack_padded_sequence(embedded, src_len.cpu(), enforce_sorted=False)
        outputs, (h, c) = self.lstm(packed)
        return h, c


class Decoder(nn.Module):
    def __init__(self, vocab, emb_dim, hid_dim, n_layers, dropout):
        super().__init__()
        self.embedding = nn.Embedding(vocab, emb_dim)
        self.lstm = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout)
        self.fc = nn.Linear(hid_dim, vocab)

    def forward(self, input, h, c):
        input = input.unsqueeze(0)
        embedded = self.embedding(input)
        output, (h, c) = self.lstm(embedded, (h, c))
        pred = self.fc(output.squeeze(0))
        return pred, h, c
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

ENC_EMB = DEC_EMB = 256
HID_DIM = 512
LAYERS = 2
DROPOUT = 0.3
LR = 1e-3
EPOCHS = 10

enc = Encoder(len(src_stoi), ENC_EMB, HID_DIM, LAYERS, DROPOUT).to(device)
dec = Decoder(len(trg_stoi), DEC_EMB, HID_DIM, LAYERS, DROPOUT).to(device)

criterion = nn.CrossEntropyLoss(ignore_index=trg_stoi[PAD])
optimizer = optim.Adam(list(enc.parameters()) + list(dec.parameters()), lr=LR)

train_losses = []
val_losses = []
```

```python
class Encoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim, n_layers, dropout, pad_idx):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_dim, padding_idx=pad_idx)
        self.lstm = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout, batch_first=True)

    def forward(self, src, src_lens):
        emb = self.emb(src)
        packed = nn.utils.rnn.pack_padded_sequence(emb, src_lens.cpu(), batch_first=True, enforce_sorted=False)
        _, (h,c) = self.lstm(packed)
        return h, c

class Decoder(nn.Module):
    def __init__(self, vocab_size, emb_dim, hid_dim, n_layers, dropout, pad_idx):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_dim, padding_idx=pad_idx)
        self.lstm = nn.LSTM(emb_dim, hid_dim, n_layers, dropout=dropout, batch_first=True)
        self.fc = nn.Linear(hid_dim, vocab_size)

    def forward(self, input, hidden, cell):
        input = input.unsqueeze(1)     # [batch → batch,1]
        emb = self.emb(input)
        out, (h,c) = self.lstm(emb, (hidden,cell))
        pred = self.fc(out.squeeze(1))
        return pred, h, c

class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device, trg_pad_idx):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device  = device
        self.pad_idx = trg_pad_idx

    def forward(self, src, src_lens, trg, teacher_forcing=0.5):
        batch, trg_len = trg.shape
        vocab = self.decoder.fc.out_features
        outputs = torch.zeros(batch, trg_len, vocab).to(self.device)

        hidden, cell = self.encoder(src, src_lens)
        input_tok = trg[:,0]

        for t in range(1, trg_len):
            output, hidden, cell = self.decoder(input_tok, hidden, cell)
            outputs[:, t] = output

            teacher = random.random() < teacher_forcing
            input_tok = trg[:, t] if teacher else output.argmax(1)

        return outputs
```
